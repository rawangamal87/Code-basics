```python
import math
```

```python
# calculates the sigmoid function of a given input x. The sigmoid function is a mathematical function that maps any real number to a value between 0 and 1
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(10)
```

```python
sigmoid(-30)
```

```python
#calculates the hyperbolic tangent function of a given input x. The hyperbolic tangent function is a mathematical function that maps any real number to a value between -1 and 1.
def tanh(x):
    return (math.exp(x)-math.exp(-x)) / (math.exp(x)) + math.exp(-x)
```

```python
tanh(100)
```

```python
tanh(10)
```

```python
tanh(-10)
```

```python
#. The ReLU function is a mathematical function that returns the input if it's positive and 0 otherwise
def relu(x):
    return max(0, x)
```

```python
relu(100)
```

```python
relu(10)
```

```python
relu(1)
```

```python
relu(-30)
```

```python
relu(-1)
```
