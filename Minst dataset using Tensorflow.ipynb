```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load the data
 (x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
#flattten the data using reshape function
x_train_flatten = x_train.reshape(len(x_train), 28*28)
```

```python
x_train_flatten.shape
```

```python
#flattten the data using reshape function
x_test_flatten = x_test.reshape(len(x_test), 28*28)
```

```python
x_test_flatten.shape
```

```python
#Sequential means the stack of layers in my neural network
#dense means all neurons
model = keras.Sequential([
         keras.layers.Dense(10, input_shape=(784,),activation='sigmoid')

])
```

```python
#pass a bunch of arrguments
#optimizer training when backward propagation
model.compile(optimizer= 'adam', loss= 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#epochs mean like number of itration
model.fit(x_train_flatten, y_train, epochs=10)
```

```python
model.evaluate(x_test_flatten, y_test)
```

```python
y_predicted = model.predict(x_test_flatten)
y_predicted[4]
```

```python
#look at the maxiumam value
y_predicted_labels = np.argmax(y_predicted[4])
```

```python
# Calculate the predicted labels for all samples
y_predicted_labels = [np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_labels)
cm
```

```python
y_test[:5]
```

```python
import seaborn as sns
plt.figure(figsize = (15, 7))
sns.heatmap(cm, annot = True, fmt = 'd')
plt.ylabel('Actual')
plt.xlabel('predicted')
```
